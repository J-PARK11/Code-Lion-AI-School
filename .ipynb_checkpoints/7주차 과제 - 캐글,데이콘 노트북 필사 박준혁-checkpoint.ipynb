# 7th Week : Bike Sharing Demand
[Data Information]   
Data Source: https://www.kaggle.com/competitions/bike-sharing-demand/overview   
Data Name : bike-sharing-demand/train.csv, test.csv   
Data Format : (10886,12), (6493,9)

[Description]   
You are provided hourly rental data spanning two years. For this competition, the training set is comprised of the first 19 days of each month, while the test set is the 20th to the end of the month. You must predict the total count of bikes rented during each hour covered by the test set, using only information available prior to the rental period.

[Attribute]   
- datetime : hourly date + timestamp  
- season :  1 = spring, 2 = summer, 3 = fall, 4 = winter 
- holiday : whether the day is considered a holiday
- workingday : whether the day is neither a weekend nor holiday
- weather : 1,2,3,4 by weather of the day (Clear, Mist, Light Rain, Heavy Rain)
- temp : temperature in Celsius
- atemp : "feels like" temperature in Celsius
- humidity : relative humidity
- windspeed : wind speed
- casual : number of non-registered user rentals initiated
- registered : number of registered user rentals initiated
- count : number of total rentals

### 그래프 색상 그라데이션 설정

```python
def gradientbars(bars,cmap):
    grad = np.expand_dims(np.linspace(0,1,256),0).T
    ax = bars[0].axes 
    lim = ax.get_xlim()+ax.get_ylim() 
    for bar in bars:
        bar.set_facecolor("none") 
        x,y = bar.get_xy() 
        w, h = bar.get_width(), bar.get_height() 
        ax.imshow(grad, extent=[x,x+w,y+h,y], cmap = cmap, aspect="auto",\
                   vmin=-0.5, vmax=2) 
    ax.axis(lim)
```

### 사용자 정의 함수

```python
def get_rmsle(y_actual,y_pred):
    diff = np.log(y_pred + 1) - np.log(y_actual + 1)
    mean_error = np.square(diff).mean()
    return np.sqrt(mean_error)
    
    
def outlier_process(x):
    upper, lower= np.quantile(x,0.75), np.quantile(x,0.25)
    iqr = upper-lower
    upper_line, lower_line = upper + iqr*1.5, lower - iqr*1.5
    print(x.name,'boundary :',np.round((lower_line,upper_line),4))
    
    x[x.values > upper_line] = np.nan
    x[x.values < lower_line] = np.nan
    x.interpolate(method="ffill", inplace=True)
    x.interpolate(method="bfill", inplace=True)
    
    return x
```

### Outlier Detection & Processing

```python
train['humidity'] = outlier_process(train['humidity'])
train['windspeed'] = outlier_process(train['windspeed'])
```

```python
test['windspeed'] = outlier_process(test['windspeed'])
```

### Feature Engineering

```python
train_df = train.drop(['casual','registered'],axis=1)
train_df['datetime']=pd.to_datetime(train_df['datetime'])
train_df['year'] = train_df['datetime'].dt.year
train_df['month'] = train_df['datetime'].dt.month
train_df['hour'] = train_df['datetime'].dt.hour
train_df.index = train_df['datetime']
train_df.drop('datetime',axis=1,inplace=True)
train_df['temp_diff'] = train_df['temp']-train_df['atemp']

print('train_df shape :',train_df.shape)
train_df.head(1)
```

```python
test_df = test.copy()
test_df['datetime']=pd.to_datetime(test_df['datetime'])
test_df['year'] = test_df['datetime'].dt.year
test_df['month'] = test_df['datetime'].dt.month
test_df['hour'] = test_df['datetime'].dt.hour
test_df.index = test_df['datetime']
test_df.drop('datetime',axis=1,inplace=True)
test_df['temp_diff'] = test_df['temp']-test_df['atemp']

print('test_df shape :',test_df.shape)
test_df.head(1)
```

### Data Split

```python
X = train_df.drop('count',axis=1)
y = train_df['count']
X_test = test_df.copy()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size=0.2,stratify=X['month'])

print('X_train :',X_train.shape)
print('y_train :',y_train.shape)
print('X_valid :',X_valid.shape)
print('y_valid :',y_valid.shape)
print('X_test :',X_test.shape)
```

## 3. Modeling

```python
lgbm_clf = LGBMRegressor(n_estimators=300)
evals = [(X_valid,y_valid)]

param_grid = {
    'learning_rate':[0.05,0.1,0.2,0.3],
    'subsample':[0.7,0.8,0.9],
    'max_depth':[7,8,9,10,11,12,13,14]
}

start_time = time.time()
grid_cv = GridSearchCV(lgbm_clf, cv=3, refit=True, param_grid=param_grid)
grid_cv.fit(X_train, y_train, early_stopping_rounds=50, eval_metric='rmsle', eval_set = evals)
end_time = time.time()
```

```python
duration = end_time - start_time
result = dt.timedelta(seconds=duration)
print('Training Time :',str(result).split('.')[0])
```

```python
print('best HyperParameter :',grid_cv.best_params_)
print('best RMSLE Score :',np.round(grid_cv.best_score_,4))
```

```python
best_est = grid_cv.best_estimator_
```

```python
feature_imp = pd.Series(data = best_est.feature_importances_, index = X.columns).sort_values(ascending=False)
feature_imp
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=feature_imp.values, y=feature_imp.index)
plt.title('Feature Importances')
plt.show()
```

## 4. Performance Evaluation

```python
y_valid_pred = best_est.predict(X_valid)
y_valid_pred = pd.Series(index = y_valid.index, data = y_valid_pred)
result = y_valid.to_frame().rename(columns={'count':'Truth'})

result = pd.concat([result,y_valid_pred],axis=1).rename(columns={0:'Pred'})
result['Error'] = np.round(np.abs(result['Truth']-result['Pred']),4)
print("result.shape :",result.shape)

result.head()
```

```python
get_rmsle(result['Truth'],result['Pred'])
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(result.index, result['Truth'], alpha=0.5, label='Truth')
sns.lineplot(result.index, result['Pred'], alpha=0.5, label='Pred')
plt.title('Truth and Pred Coincidences',size=20)
plt.legend(loc='upper right')
plt.show()
```

```python
upper, lower= np.quantile(result['Error'].values,0.75), np.quantile(result['Error'].values,0.25)
iqr = upper-lower
boundary = upper + iqr*1.5
print('boundary :',np.round(boundary,4))
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(result.index, result['Error'], color = 'red', alpha=0.5, marker='*', label='Error Points')
sns.lineplot(result.index, boundary, color='blue', linestyle='--', label='Outlier Boundary', alpha=0.8)
plt.title('Prediction Error Variances')
plt.legend(loc="upper right")
plt.show()
```
